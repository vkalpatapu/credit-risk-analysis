**Dataset:** Kaggle Credit Risk Dataset | 32,581 rows | 12 columns

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('credit_risk_dataset.csv')
print('Shape:', df.shape)
print(df.head())
```

```python
# Check missing values
print(df.isnull().sum())

# Fill missing employment length with median
df['person_emp_length'].fillna(df['person_emp_length'].median(), inplace=True)
print('Missing values after fix:', df.isnull().sum().sum())
```

```python
# Overall default rate
total = len(df)
defaults = df['loan_status'].sum()
default_rate = defaults / total * 100
print(f'Total Loans: {total:,}')
print(f'Total Defaults: {defaults:,}')
print(f'Default Rate: {default_rate:.2f}%')
```

```python
# Default rate by loan grade
grade_default = df.groupby('loan_grade')['loan_status'].mean() * 100
print(grade_default.sort_values(ascending=False).round(2))
print('\nKey insight: Risk cliff between Grade C (20%) and Grade D (59%)')
```

```python
# Default rate by grade - Bar chart
grade_default.sort_values(ascending=False).plot(kind='bar', color='coral', figsize=(8,5))
plt.title('Default Rate by Loan Grade')
plt.ylabel('Default Rate (%)')
plt.xlabel('Loan Grade')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
# Default rate by income bracket
df['income_bracket'] = pd.cut(df['person_income'],
    bins=[0, 30000, 60000, 100000, float('inf')],
    labels=['Low Income', 'Middle Income', 'Upper Middle', 'High Income'])

income_default = df.groupby('income_bracket')['loan_status'].mean() * 100
print(income_default.sort_values(ascending=False).round(2))
print('\nKey insight: Low income 47% default vs High income 8.93%')
```
